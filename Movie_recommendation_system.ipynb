```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies=movies.drop(columns=['id','budget','homepage','original_language','original_title','popularity','production_companies','revenue','runtime','spoken_languages','status','vote_average','vote_count'])
```

```python
movies=movies.drop(columns=['release_date','production_countries'])
```

```python
movies
```

```python
id=movies['movie_id']
movies=movies.drop(columns=['movie_id','tagline'])
```

```python
movies
```

```python
import ast
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert_cast(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter < 3:
            L.append(i['name'])
            counter +=1
    return L
        
```

```python
movies['cast']=movies['cast'].apply(convert_cast)
```

```python
movies
```

```python
movies['crew'][0]
```

```python
def fetch_dir(crew):
    L=[]
    for i in ast.literal_eval(crew):
        if i['job']== "Director" :
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_dir)
```

```python
movies['id']=id
```

```python
def join(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(join)
movies['crew'] = movies['crew'].apply(join)
movies['genres'] = movies['genres'].apply(join)
movies['keywords'] = movies['keywords'].apply(join)
```

```python
movies
```

```python
movies['overview']
```

```python
movies['overview']=movies['overview'].astype(str).apply(lambda x:x.split())
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies=movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(movies['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```
